```python
import pandas as pd
import seaborn as sns
```

```python
airline = pd.read_csv("airline-safety.csv")
airline.head()
```

* airline	Airline (asterisk indicates that regional subsidiaries are included)
* avail_seat_km_per_week: Available seat kilometers flown every week
* incidents_85_99:	Total number of incidents, 1985–1999
* fatal_accidents_85_99:	Total number of fatal accidents, 1985–1999
* fatalities_85_99:	Total number of fatalities, 1985–1999
* incidents_00_14:	Total number of incidents, 2000–2014
* fatal_accidents_00_14:	Total number of fatal accidents, 2000–2014
* fatalities_00_14:	Total number of fatalities, 2000–2014

```python
airline.info()
```

```python
airline.describe(include='all')
```

* We have 56 airline companies 
* we notice that some max values are a lot bigger than their corresponding 75% values, hence the existance of outliers (see *incidents_85_99*)


```python
# plotting the airlines in function to incidents_85_99
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.scatterplot(data=airline, x='airline', y='incidents_85_99')
ax.tick_params(axis='x', rotation=90)
# the outlier in this case is Aeroflot
```

```python
# let's see the relationship between the number of incidents and the fatalities
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.lineplot(data=airline, x='incidents_85_99', y='fatalities_85_99')
ax.tick_params(axis='x', rotation=90)
# we notice that a bigger number of incidents does not mean a higher number of fatalities, the mean is between 100 and 200
```

```python
# let's see the relationship between the number of fatal_accidents and the fatalities
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.lineplot(data=airline, x='fatal_accidents_85_99', y='fatalities_85_99')
ax.tick_params(axis='x', rotation=90)
# the number of fatalities keeps on increasing as the number of fatal accidents increased, except for cases where the number of fatal accidents is 7 or more
```

```python
# plotting the avail_seat_km_per_week in function to incidents_85_99
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.scatterplot(data=airline, x='avail_seat_km_per_week', y='incidents_85_99')
ax.tick_params(axis='x', rotation=90)
# in this case, apart from the outlier "Aeroflot", most of the airlines make a cluster where the lower the number of avail_seat_km_per_week is, the lower the number of incidents 85-99
```

```python
# plotting the avail_seat_km_per_week in function to incidents_00_14
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.scatterplot(data=airline, x='avail_seat_km_per_week', y='incidents_00_14')
ax.tick_params(axis='x', rotation=90)
# in this case, we have three outliers, let's find them
```

```python
airline[airline.avail_seat_km_per_week > 5e9] 
# these are the outliers in incidents 2000 and 2014
# although 
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
ax = sns.lineplot(data=airline, x='fatal_accidents_00_14', y='fatalities_00_14')
ax.tick_params(axis='x', rotation=90)
# as expected, the number of fata accidents is correlated with the number of fatalities 2000-2014
```

**Let's find the most safe airline**

```python
# let's find the airline that had no fatalities at all
airline[(airline.fatalities_00_14 == 0) & (airline.fatalities_85_99 == 0)]
# here are the airlines that although had incidents, there are no recorded fatalities
```

```python
# among these airlines, which are the ones with no incidents at all
airline[(airline.fatalities_00_14 == 0) & (airline.fatalities_85_99 == 0) & (airline.incidents_85_99 == 0) & (airline.incidents_00_14 == 0)]
# the one that verifies these conditions is TAP - Air Portugal
# however, the reason for that is because of the small number of avail_seat_km_per_week, hence it is not totally guaranteed that incidents wont happen in the future
```

```python
# let's redo our conditioning, but this time, let's take into consideration the avail_seat_km_per_week parameter, and instead, let's add thresholds to the number of incidents
# let's look at the dataset stats again
airline.describe(include='all')
```

As thresholds for the conditions, we will take the quantile of the dataset, which are: <br>
* 4 for incidents 85-99
* 3 for incidents 00-14

```python
airline[(airline.fatalities_00_14 == 0) & (airline.fatalities_85_99 == 0) & (airline.incidents_85_99 <= 4) & (airline.incidents_00_14 <= 3)]
# from these we take the ones with high avail_seat_km_per_week
```

**the airlines we choose are:**
* Air Canada
* Cathay Pacific
* Virgin Atlantic
